** Do the Train Test Split** using test size 0.3

```python
# Code here please
from sklearn.model_selection import train_test_split

# Prepare features (X) and target (y)
# Drop the Target column from features
X = df.drop('Target', axis=1)  # Features: BI-RADS, Age, Shape, Margin, Density
y = df['Target']  # Target: 0=Benign, 1=Malignant

# Perform train-test split (70% train, 30% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Check the split results
print("=" * 50)
print("TRAIN-TEST SPLIT RESULTS")
print("=" * 50)
print(f"Total samples: {len(df)}")
print(f"Training set size: {len(X_train)} samples ({len(X_train)/len(df)*100:.1f}%)")
print(f"Test set size: {len(X_test)} samples ({len(X_test)/len(df)*100:.1f}%)")

print("\n" + "=" * 50)
print("TARGET DISTRIBUTION")
print("=" * 50)
print("\nOriginal dataset:")
print(y.value_counts())
print(f"Benign: {sum(y==0)} ({sum(y==0)/len(y)*100:.1f}%)")
print(f"Malignant: {sum(y==1)} ({sum(y==1)/len(y)*100:.1f}%)")

print("\nTraining set:")
print(y_train.value_counts())
print(f"Benign: {sum(y_train==0)} ({sum(y_train==0)/len(y_train)*100:.1f}%)")
print(f"Malignant: {sum(y_train==1)} ({sum(y_train==1)/len(y_train)*100:.1f}%)")

print("\nTest set:")
print(y_test.value_counts())
print(f"Benign: {sum(y_test==0)} ({sum(y_test==0)/len(y_test)*100:.1f}%)")
print(f"Malignant: {sum(y_test==1)} ({sum(y_test==1)/len(y_test)*100:.1f}%)")

# Optional: Verify that the split preserved the class distribution
print("\n" + "=" * 50)
print("CLASS DISTRIBUTION COMPARISON")
print("=" * 50)
print(f"Original - Malignant rate: {y.mean()*100:.1f}%")
print(f"Training - Malignant rate: {y_train.mean()*100:.1f}%")
print(f"Test - Malignant rate: {y_test.mean()*100:.1f}%")
```

```python
# Code here please
# Example: Train a Decision Tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report

# Create and train the model
dt_model = DecisionTreeClassifier(random_state=42)
dt_model.fit(X_train, y_train)

# Make predictions
y_pred = dt_model.predict(X_test)

# Evaluate
print("\n" + "=" * 50)
print("DECISION TREE RESULTS")
print("=" * 50)
print(f"Accuracy: {accuracy_score(y_test, y_pred):.3f}")
print("\nClassification Report:")
print(classification_report(y_test, y_pred, target_names=['Benign', 'Malignant']))
```

```python
# Assuming you've already done the train-test split
# If not, run this first:
from sklearn.model_selection import train_test_split

X = df.drop('Target', axis=1)
y = df['Target']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Import and create Decision Tree instance
from sklearn.tree import DecisionTreeClassifier

# Create the model (using tuned parameters for this dataset)
dt_model = DecisionTreeClassifier(
    criterion='entropy',
    max_depth=4,
    min_samples_split=20,
    min_samples_leaf=10,
    random_state=42
)

# FIT THE MODEL TO TRAINING DATA
print("Training the Decision Tree model...")
dt_model.fit(X_train, y_train)

# Check if model was fitted successfully
print("✓ Model training complete!")
print(f"\nTraining set size: {len(X_train)} samples")
print(f"Number of features used: {X_train.shape[1]}")

# Quick check of model parameters after fitting
print(f"\nTree depth: {dt_model.get_depth()}")
print(f"Number of leaves: {dt_model.get_n_leaves()}")

# Optional: Check feature importance
feature_importance = pd.DataFrame({
    'feature': X.columns,
    'importance': dt_model.feature_importances_
}).sort_values('importance', ascending=False)

print("\nFeature Importance:")
print(feature_importance)

# Make predictions on training set to check performance
train_pred = dt_model.predict(X_train)
train_accuracy = (train_pred == y_train).mean()
print(f"\nTraining accuracy: {train_accuracy:.3f}")

# Make predictions on test set
test_pred = dt_model.predict(X_test)
test_accuracy = (test_pred == y_test).mean()
print(f"Test accuracy: {test_accuracy:.3f}")

# Check for overfitting
if train_accuracy - test_accuracy > 0.1:
    print("\n⚠️ Warning: Possible overfitting (train accuracy much higher than test accuracy)")
    print("Consider reducing max_depth or increasing min_samples_split/leaf")
else:
    print("\n✓ Model generalizes well (similar train and test accuracy)")
```

```python
# Make predictions on test set
y_pred = dt_model.predict(X_test)

# Get prediction probabilities (useful for medical decisions)
y_pred_proba = dt_model.predict_proba(X_test)

# Show first 5 predictions with probabilities
print("\nSample Predictions (first 5 test samples):")
for i in range(5):
    pred_class = "Malignant" if y_pred[i] == 1 else "Benign"
    actual_class = "Malignant" if y_test.iloc[i] == 1 else "Benign"
    prob_benign = y_pred_proba[i][0] * 100
    prob_malignant = y_pred_proba[i][1] * 100
    print(f"\nSample {i+1}:")
    print(f"  Predicted: {pred_class} (Benign: {prob_benign:.1f}%, Malignant: {prob_malignant:.1f}%)")
    print(f"  Actual: {actual_class}")
    print(f"  Correct: {'✓' if y_pred[i] == y_test.iloc[i] else '✗'}")
```

**What is the Score of the model?**

```python
# Code here please
# Get the model score (accuracy) on training data
train_score = dt_model.score(X_train, y_train)
print(f"Training Accuracy Score: {train_score:.4f} ({train_score*100:.2f}%)")

# Get the model score (accuracy) on test data
test_score = dt_model.score(X_test, y_test)
print(f"Test Accuracy Score: {test_score:.4f} ({test_score*100:.2f}%)")

# Calculate the difference
score_diff = train_score - test_score
print(f"\nDifference (Train - Test): {score_diff:.4f} ({score_diff*100:.2f}%)")

# Interpretation
print("\n" + "=" * 50)
print("MODEL PERFORMANCE INTERPRETATION")
print("=" * 50)

if test_score >= 0.85:
    print("✓ EXCELLENT: Model is performing very well (>85% accuracy)")
elif test_score >= 0.75:
    print("✓ GOOD: Model is performing well (75-85% accuracy)")
elif test_score >= 0.65:
    print("✓ FAIR: Model is performing decently (65-75% accuracy)")
else:
    print("⚠️ POOR: Model needs improvement (<65% accuracy)")

if score_diff > 0.1:
    print("⚠️ OVERFITTING: Model performs much better on training data")
    print("   Consider: reducing max_depth, increasing min_samples_split/leaf")
elif score_diff < -0.05:
    print("⚠️ UNUSUAL: Test accuracy is higher than training accuracy")
    print("   This might indicate data leakage or an unusual split")
else:
    print("✓ GOOD: Model generalizes well (small train-test gap)")

# Also get more detailed metrics
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

y_pred = dt_model.predict(X_test)

print("\n" + "=" * 50)
print("DETAILED METRICS (Test Set)")
print("=" * 50)
print(f"Accuracy:  {accuracy_score(y_test, y_pred):.4f}")
print(f"Precision: {precision_score(y_test, y_pred):.4f} (Of those predicted malignant, how many actually were?)")
print(f"Recall:    {recall_score(y_test, y_pred):.4f} (Of actual malignant cases, how many did we catch?)")
print(f"F1-Score:  {f1_score(y_test, y_pred):.4f} (Harmonic mean of precision and recall)")

# Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)

print("\n" + "=" * 50)
print("CONFUSION MATRIX")
print("=" * 50)
print("               Predicted")
print("               Benign  Malignant")
print(f"Actual Benign    {cm[0,0]:>3}      {cm[0,1]:>3}")
print(f"       Malignant {cm[1,0]:>3}      {cm[1,1]:>3}")
print("\nInterpretation:")
print(f"- True Negatives (Correct Benign): {cm[0,0]}")
print(f"- False Positives (Incorrect Malignant): {cm[0,1]}")
print(f"- False Negatives (Missed Cancer): {cm[1,0]}")
print(f"- True Positives (Correct Malignant): {cm[1,1]}")
```
