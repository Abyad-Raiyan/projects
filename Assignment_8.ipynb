````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('bollywood_complete_dataset.csv')

# Calculate ROI if not already present
df['ROI'] = ((df['BoxOfficeCollection'] - df['Budget']) / df['Budget'] * 100).round(2)

# Filter Comedy and Drama genres
comedy_df = df[df['Genre'] == 'Comedy']
drama_df = df[df['Genre'] == 'Drama']

print("=" * 60)
print("COMEDY vs DRAMA ROI COMPARISON")
print("=" * 60)

# Basic statistics
print("\nCOMEDY GENRE:")
print(f"Number of movies: {len(comedy_df)}")
print(f"Mean ROI: {comedy_df['ROI'].mean():.2f}%")
print(f"Median ROI: {comedy_df['ROI'].median():.2f}%")
print(f"Std Dev: {comedy_df['ROI'].std():.2f}%")
print(f"Min ROI: {comedy_df['ROI'].min():.2f}%")
print(f"Max ROI: {comedy_df['ROI'].max():.2f}%")

print("\nDRAMA GENRE:")
print(f"Number of movies: {len(drama_df)}")
print(f"Mean ROI: {drama_df['ROI'].mean():.2f}%")
print(f"Median ROI: {drama_df['ROI'].median():.2f}%")
print(f"Std Dev: {drama_df['ROI'].std():.2f}%")
print(f"Min ROI: {drama_df['ROI'].min():.2f}%")
print(f"Max ROI: {drama_df['ROI'].max():.2f}%")

# Visualization
fig, axes = plt.subplots(2, 2, figsize=(15, 12))
fig.suptitle('Comedy vs Drama ROI Distribution', fontsize=16, fontweight='bold')

# 1. Histogram comparison
axes[0, 0].hist(comedy_df['ROI'].dropna(), bins=15, alpha=0.7, label='Comedy', color='skyblue', edgecolor='black')
axes[0, 0].hist(drama_df['ROI'].dropna(), bins=15, alpha=0.7, label='Drama', color='lightcoral', edgecolor='black')
axes[0, 0].set_xlabel('ROI (%)', fontsize=12)
axes[0, 0].set_ylabel('Frequency', fontsize=12)
axes[0, 0].set_title('Histogram Comparison', fontsize=14)
axes[0, 0].legend()
axes[0, 0].axvline(comedy_df['ROI'].mean(), color='blue', linestyle='dashed', linewidth=2, label=f'Comedy Mean: {comedy_df["ROI"].mean():.1f}%')
axes[0, 0].axvline(drama_df['ROI'].mean(), color='red', linestyle='dashed', linewidth=2, label=f'Drama Mean: {drama_df["ROI"].mean():.1f}%')
axes[0, 0].legend()

# 2. KDE plot
sns.kdeplot(data=comedy_df, x='ROI', ax=axes[0, 1], label='Comedy', fill=True, alpha=0.5, color='skyblue')
sns.kdeplot(data=drama_df, x='ROI', ax=axes[0, 1], label='Drama', fill=True, alpha=0.5, color='lightcoral')
axes[0, 1].set_xlabel('ROI (%)', fontsize=12)
axes[0, 1].set_ylabel('Density', fontsize=12)
axes[0, 1].set_title('Distribution Plot (KDE) Comparison', fontsize=14)
axes[0, 1].legend()

# 3. Box plot comparison
data_to_plot = [comedy_df['ROI'].dropna(), drama_df['ROI'].dropna()]
bp = axes[1, 0].boxplot(data_to_plot, labels=['Comedy', 'Drama'], patch_artist=True)
bp['boxes'][0].set_facecolor('skyblue')
bp['boxes'][1].set_facecolor('lightcoral')
axes[1, 0].set_ylabel('ROI (%)', fontsize=12)
axes[1, 0].set_title('Box Plot Comparison', fontsize=14)
axes[1, 0].grid(True, alpha=0.3)

# 4. Bar chart comparison
genres = ['Comedy', 'Drama']
means = [comedy_df['ROI'].mean(), drama_df['ROI'].mean()]
medians = [comedy_df['ROI'].median(), drama_df['ROI'].median()]
x = np.arange(len(genres))
width = 0.35

axes[1, 1].bar(x - width/2, means, width, label='Mean ROI', color='skyblue', edgecolor='black')
axes[1, 1].bar(x + width/2, medians, width, label='Median ROI', color='lightgreen', edgecolor='black')
axes[1, 1].set_xlabel('Genre', fontsize=12)
axes[1, 1].set_ylabel('ROI (%)', fontsize=12)
axes[1, 1].set_title('Mean vs Median ROI Comparison', fontsize=14)
axes[1, 1].set_xticks(x)
axes[1, 1].set_xticklabels(genres)
axes[1, 1].legend()
axes[1, 1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Statistical test (t-test) to check if difference is significant
try:
    from scipy import stats
    t_stat, p_value = stats.ttest_ind(comedy_df['ROI'].dropna(), drama_df['ROI'].dropna())

    print("\n" + "=" * 60)
    print("STATISTICAL SIGNIFICANCE TEST")
    print("=" * 60)
    print(f"T-statistic: {t_stat:.4f}")
    print(f"P-value: {p_value:.4f}")

    if p_value < 0.05:
        print("✓ The difference in ROI between Comedy and Drama is statistically significant (p < 0.05)")
    else:
        print("✗ The difference in ROI between Comedy and Drama is NOT statistically significant (p >= 0.05)")
except:
    print("\n(Install scipy for statistical test: pip install scipy)")

# Conclusion
print("\n" + "=" * 60)
print("CONCLUSION")
print("=" * 60)

if comedy_df['ROI'].mean() > drama_df['ROI'].mean():
    print(f"✅ COMEDY genre typically sees higher ROI")
    print(f"   Comedy average ROI: {comedy_df['ROI'].mean():.2f}%")
    print(f"   Drama average ROI: {drama_df['ROI'].mean():.2f}%")
    print(f"   Difference: {comedy_df['ROI'].mean() - drama_df['ROI'].mean():.2f}%")
else:
    print(f"✅ DRAMA genre typically sees higher ROI")
    print(f"   Drama average ROI: {drama_df['ROI'].mean():.2f}%")
    print(f"   Comedy average ROI: {comedy_df['ROI'].mean():.2f}%")
    print(f"   Difference: {drama_df['ROI'].mean() - comedy_df['ROI'].mean():.2f}%")

# Additional insight: Success rate (movies with positive ROI)
comedy_success = len(comedy_df[comedy_df['ROI'] > 0]) / len(comedy_df) * 100
drama_success = len(drama_df[drama_df['ROI'] > 0]) / len(drama_df) * 100

print(f"\nSuccess Rate (ROI > 0%):")
print(f"   Comedy: {comedy_success:.1f}%")
print(f"   Drama: {drama_success:.1f}%")
````

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('bollywood_complete_dataset.csv')

# Calculate correlation
correlation = df['BoxOfficeCollection'].corr(df['YoutubeLikes'])

print("=" * 60)
print("CORRELATION ANALYSIS: Box Office Collection vs YouTube Likes")
print("=" * 60)
print(f"Correlation coefficient: {correlation:.4f}")

# Interpret correlation strength
print("\nINTERPRETATION:")
if correlation > 0.7:
    print(f"✓ Strong positive correlation ({correlation:.2f})")
    print("  As box office collection increases, YouTube likes strongly increase")
elif correlation > 0.3:
    print(f"✓ Moderate positive correlation ({correlation:.2f})")
    print("  As box office collection increases, YouTube likes tend to increase")
elif correlation > 0:
    print(f"✓ Weak positive correlation ({correlation:.2f})")
    print("  Slight tendency for YouTube likes to increase with box office")
elif correlation == 0:
    print("✗ No correlation")
    print("  No relationship between box office and YouTube likes")
elif correlation > -0.3:
    print(f"✗ Weak negative correlation ({correlation:.2f})")
    print("  Slight tendency for YouTube likes to decrease as box office increases")
elif correlation > -0.7:
    print(f"✗ Moderate negative correlation ({correlation:.2f})")
    print("  As box office increases, YouTube likes tend to decrease")
else:
    print(f"✗ Strong negative correlation ({correlation:.2f})")
    print("  As box office increases, YouTube likes strongly decrease")

# Visualization
fig, axes = plt.subplots(2, 2, figsize=(15, 12))
fig.suptitle('Box Office Collection vs YouTube Likes Correlation', fontsize=16, fontweight='bold')

# 1. Scatter plot with regression line
sns.scatterplot(data=df, x='BoxOfficeCollection', y='YoutubeLikes', ax=axes[0, 0], alpha=0.6, color='skyblue')
sns.regplot(data=df, x='BoxOfficeCollection', y='YoutubeLikes', ax=axes[0, 0], scatter=False, color='red')
axes[0, 0].set_xlabel('Box Office Collection (Cr)', fontsize=12)
axes[0, 0].set_ylabel('YouTube Likes', fontsize=12)
axes[0, 0].set_title(f'Scatter Plot with Regression Line (r = {correlation:.4f})', fontsize=14)
axes[0, 0].grid(True, alpha=0.3)

# 2. Heatmap of correlations
corr_matrix = df[['BoxOfficeCollection', 'YoutubeLikes', 'Budget', 'YoutubeViews']].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0, ax=axes[0, 1],
            square=True, linewidths=1, cbar_kws={"shrink": 0.8})
axes[0, 1].set_title('Correlation Heatmap', fontsize=14)

# 3. Joint plot style (hexbin)
hb = axes[1, 0].hexbin(df['BoxOfficeCollection'], df['YoutubeLikes'], gridsize=20, cmap='YlOrRd')
axes[1, 0].set_xlabel('Box Office Collection (Cr)', fontsize=12)
axes[1, 0].set_ylabel('YouTube Likes', fontsize=12)
axes[1, 0].set_title('Hexbin Plot (Density View)', fontsize=14)
plt.colorbar(hb, ax=axes[1, 0])

# 4. Box plot by quartiles
df['BoxOffice_Quartile'] = pd.qcut(df['BoxOfficeCollection'], q=4, labels=['Q1 (Lowest)', 'Q2', 'Q3', 'Q4 (Highest)'])
sns.boxplot(data=df, x='BoxOffice_Quartile', y='YoutubeLikes', ax=axes[1, 1], palette='viridis')
axes[1, 1].set_xlabel('Box Office Collection Quartiles', fontsize=12)
axes[1, 1].set_ylabel('YouTube Likes', fontsize=12)
axes[1, 1].set_title('YouTube Likes by Box Office Quartiles', fontsize=14)
axes[1, 1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()

# Additional statistics
print("\n" + "=" * 60)
print("ADDITIONAL INSIGHTS")
print("=" * 60)

# Top 10 movies by box office and their YouTube likes
top_10_box = df.nlargest(10, 'BoxOfficeCollection')[['MovieName', 'BoxOfficeCollection', 'YoutubeLikes']]
print("\nTop 10 Box Office Movies:")
print(top_10_box.to_string(index=False))

# Top 10 movies by YouTube likes
top_10_likes = df.nlargest(10, 'YoutubeLikes')[['MovieName', 'YoutubeLikes', 'BoxOfficeCollection']]
print("\nTop 10 YouTube Likes Movies:")
print(top_10_likes.to_string(index=False))

# Check if top movies overlap
top_box_movies = set(top_10_box['MovieName'])
top_likes_movies = set(top_10_likes['MovieName'])
overlap = top_box_movies.intersection(top_likes_movies)

print(f"\nMovies in both top 10 lists: {len(overlap)}")
if len(overlap) > 0:
    for movie in overlap:
        print(f"  • {movie}")

# Final conclusion
print("\n" + "=" * 60)
print("FINAL CONCLUSION")
print("=" * 60)
if correlation > 0:
    print(f"✅ The correlation is POSITIVE ({correlation:.4f})")
    print("   Higher box office collections tend to have more YouTube likes")
else:
    print(f"❌ The correlation is NEGATIVE ({correlation:.4f})")
    print("   Higher box office collections tend to have fewer YouTube likes")
````

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('bollywood_complete_dataset.csv')

# Calculate average YouTube likes by genre
genre_likes = df.groupby('Genre')['YoutubeLikes'].agg(['mean', 'median', 'count', 'sum']).round(2)
genre_likes = genre_likes.sort_values('mean', ascending=False)

print("=" * 70)
print("YOUTUBE LIKES BY GENRE")
print("=" * 70)
print(genre_likes)

# Find genre with highest average YouTube likes
top_genre = genre_likes.index[0]
top_mean = genre_likes.iloc[0]['mean']
top_count = genre_likes.iloc[0]['count']

print("\n" + "=" * 70)
print(f"Genre with most YouTube likes: {top_genre}")
print(f"Average YouTube likes: {top_mean:,.0f}")
print(f"Number of movies: {top_count}")

# Visualization
fig, axes = plt.subplots(2, 2, figsize=(18, 14))
fig.suptitle('YouTube Likes Distribution by Genre', fontsize=16, fontweight='bold')

# 1. Box plot for all genres
genres_order = genre_likes.index.tolist()
sns.boxplot(data=df, x='Genre', y='YoutubeLikes', order=genres_order, ax=axes[0, 0], palette='viridis')
axes[0, 0].set_xlabel('Genre', fontsize=12)
axes[0, 0].set_ylabel('YouTube Likes', fontsize=12)
axes[0, 0].set_title('Box Plot of YouTube Likes by Genre', fontsize=14)
axes[0, 0].tick_params(axis='x', rotation=45)

# 2. Violin plot (shows distribution better)
sns.violinplot(data=df, x='Genre', y='YoutubeLikes', order=genres_order, ax=axes[0, 1], palette='muted')
axes[0, 1].set_xlabel('Genre', fontsize=12)
axes[0, 1].set_ylabel('YouTube Likes', fontsize=12)
axes[0, 1].set_title('Violin Plot of YouTube Likes by Genre', fontsize=14)
axes[0, 1].tick_params(axis='x', rotation=45)

# 3. Bar chart of mean YouTube likes
colors = plt.cm.viridis(np.linspace(0, 1, len(genres_order)))
bars = axes[1, 0].bar(range(len(genres_order)), genre_likes['mean'], color=colors, edgecolor='black')
axes[1, 0].set_xlabel('Genre', fontsize=12)
axes[1, 0].set_ylabel('Average YouTube Likes', fontsize=12)
axes[1, 0].set_title('Average YouTube Likes by Genre', fontsize=14)
axes[1, 0].set_xticks(range(len(genres_order)))
axes[1, 0].set_xticklabels(genres_order, rotation=45, ha='right')
axes[1, 0].grid(True, alpha=0.3, axis='y')

# Add value labels on bars
for i, bar in enumerate(bars):
    height = bar.get_height()
    axes[1, 0].text(bar.get_x() + bar.get_width()/2., height + 0.01 * max(genre_likes['mean']),
                    f'{int(height):,}', ha='center', va='bottom', rotation=0, fontsize=9)

# 4. Top genres pie chart
top_5_genres = genre_likes.head(5)
other_count = len(df[~df['Genre'].isin(top_5_genres.index)])
other_mean = df[~df['Genre'].isin(top_5_genres.index)]['YoutubeLikes'].mean()

# Prepare data for pie
pie_labels = list(top_5_genres.index) + ['Other Genres']
pie_values = list(top_5_genres['sum']) + [df[~df['Genre'].isin(top_5_genres.index)]['YoutubeLikes'].sum()]

colors_pie = plt.cm.viridis(np.linspace(0, 1, len(pie_labels)))
axes[1, 1].pie(pie_values, labels=pie_labels, autopct='%1.1f%%', colors=colors_pie, startangle=90)
axes[1, 1].set_title('Share of Total YouTube Likes by Genre', fontsize=14)
axes[1, 1].axis('equal')

plt.tight_layout()
plt.show()

# Box plot for top 5 genres only (clearer view)
plt.figure(figsize=(14, 8))
top_5_df = df[df['Genre'].isin(top_5_genres.index[:5])]
sns.boxplot(data=top_5_df, x='Genre', y='YoutubeLikes', order=top_5_genres.index[:5], palette='Set2')
plt.xlabel('Genre', fontsize=14)
plt.ylabel('YouTube Likes', fontsize=14)
plt.title('YouTube Likes Distribution - Top 5 Genres', fontsize=16, fontweight='bold')
plt.grid(True, alpha=0.3, axis='y')
plt.tight_layout()
plt.show()

# Statistical summary for each genre
print("\n" + "=" * 70)
print("DETAILED STATISTICS BY GENRE")
print("=" * 70)

for genre in genres_order[:10]:  # Show top 10 genres
    genre_data = df[df['Genre'] == genre]['YoutubeLikes']
    print(f"\n{genre}:")
    print(f"  Count: {len(genre_data)}")
    print(f"  Mean: {genre_data.mean():,.0f}")
    print(f"  Median: {genre_data.median():,.0f}")
    print(f"  Std Dev: {genre_data.std():,.0f}")
    print(f"  Min: {genre_data.min():,.0f}")
    print(f"  Max: {genre_data.max():,.0f}")
    print(f"  Q1: {genre_data.quantile(0.25):,.0f}")
    print(f"  Q3: {genre_data.quantile(0.75):,.0f}")

# Conclusion
print("\n" + "=" * 70)
print("CONCLUSION")
print("=" * 70)
print(f"✅ Genre with highest average YouTube likes: {top_genre} ({top_mean:,.0f} likes)")

print("\nTop 5 Genres by YouTube Likes:")
for i, (genre, row) in enumerate(genre_likes.head(5).iterrows(), 1):
    print(f"  {i}. {genre}: {row['mean']:,.0f} avg likes ({row['count']} movies)")

# Check if there's a significant difference between top and bottom
bottom_genre = genre_likes.index[-1]
bottom_mean = genre_likes.iloc[-1]['mean']
print(f"\nGenre with lowest average YouTube likes: {bottom_genre} ({bottom_mean:,.0f} likes)")
print(f"Ratio (Top/Bottom): {top_mean/bottom_mean:.1f}x")
````

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('bollywood_complete_dataset.csv')

# Select numerical columns for correlation analysis
numeric_cols = ['Budget', 'BoxOfficeCollection', 'YoutubeViews', 'YoutubeLikes', 'YoutubeDislikes']
corr_df = df[numeric_cols]

# Calculate correlation matrix
corr_matrix = corr_df.corr()

print("=" * 70)
print("CORRELATION MATRIX")
print("=" * 70)
print(corr_matrix.round(4))

# 1. Heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0,
            square=True, linewidths=2, cbar_kws={"shrink": 0.8},
            annot_kws={"size": 14, "weight": "bold"})
plt.title('Correlation Heatmap of Movie Variables', fontsize=16, fontweight='bold', pad=20)
plt.tight_layout()
plt.show()

# 2. Pair plot
plt.figure(figsize=(15, 15))
g = sns.pairplot(corr_df, diag_kind='kde', plot_kws={'alpha': 0.6, 's': 80, 'edgecolor': 'k'})
g.fig.suptitle('Pair Plot of Movie Variables', fontsize=16, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()

# 3. Clustered heatmap (better for seeing patterns)
plt.figure(figsize=(12, 10))
sns.clustermap(corr_matrix, annot=True, cmap='coolwarm', center=0,
               linewidths=2, figsize=(12, 10), dendrogram_ratio=0.1,
               cbar_pos=(0.02, 0.83, 0.03, 0.15))
plt.suptitle('Clustered Correlation Heatmap', fontsize=16, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()

# Find highly correlated pairs (|correlation| > 0.7)
print("\n" + "=" * 70)
print("HIGHLY CORRELATED PAIRS (|r| > 0.7)")
print("=" * 70)

high_corr = []
for i in range(len(corr_matrix.columns)):
    for j in range(i+1, len(corr_matrix.columns)):
        corr_value = corr_matrix.iloc[i, j]
        if abs(corr_value) > 0.7:
            high_corr.append({
                'Variable 1': corr_matrix.columns[i],
                'Variable 2': corr_matrix.columns[j],
                'Correlation': round(corr_value, 4)
            })

if high_corr:
    high_corr_df = pd.DataFrame(high_corr)
    print(high_corr_df.to_string(index=False))
else:
    print("No highly correlated pairs found (|r| > 0.7)")

# Find moderately correlated pairs (0.5 < |correlation| < 0.7)
print("\n" + "=" * 70)
print("MODERATELY CORRELATED PAIRS (0.5 < |r| < 0.7)")
print("=" * 70)

mod_corr = []
for i in range(len(corr_matrix.columns)):
    for j in range(i+1, len(corr_matrix.columns)):
        corr_value = corr_matrix.iloc[i, j]
        if 0.5 < abs(corr_value) <= 0.7:
            mod_corr.append({
                'Variable 1': corr_matrix.columns[i],
                'Variable 2': corr_matrix.columns[j],
                'Correlation': round(corr_value, 4)
            })

if mod_corr:
    mod_corr_df = pd.DataFrame(mod_corr)
    print(mod_corr_df.to_string(index=False))
else:
    print("No moderately correlated pairs found")

# Summary statistics
print("\n" + "=" * 70)
print("CORRELATION SUMMARY")
print("=" * 70)

# Find highest positive correlation
max_corr = corr_matrix.unstack().sort_values(ascending=False)
max_corr = max_corr[max_corr < 1]  # Remove self-correlations
if not max_corr.empty:
    top_pair = max_corr.index[0]
    print(f"Highest positive correlation: {top_pair[0]} & {top_pair[1]} = {max_corr.iloc[0]:.4f}")

# Find highest negative correlation (if any)
min_corr = corr_matrix.unstack().sort_values()
min_corr = min_corr[min_corr < 1]  # Remove self-correlations
if not min_corr.empty and min_corr.iloc[0] < 0:
    bottom_pair = min_corr.index[0]
    print(f"Highest negative correlation: {bottom_pair[0]} & {bottom_pair[1]} = {min_corr.iloc[0]:.4f}")

# Average correlation
avg_corr = corr_matrix.values[np.triu_indices_from(corr_matrix.values, k=1)].mean()
print(f"Average correlation between variables: {avg_corr:.4f}")

# Detailed interpretation
print("\n" + "=" * 70)
print("INTERPRETATION")
print("=" * 70)

# Interpret each variable's relationship with BoxOfficeCollection
print("\nRelationship with Box Office Collection:")
for col in numeric_cols[1:]:  # Skip Budget for first comparison
    if col != 'BoxOfficeCollection':
        corr = corr_matrix.loc['BoxOfficeCollection', col]
        strength = "strong" if abs(corr) > 0.7 else "moderate" if abs(corr) > 0.5 else "weak"
        direction = "positive" if corr > 0 else "negative"
        print(f"  • {col}: {strength} {direction} correlation ({corr:.4f})")

# Interpret relationships with YouTube metrics
print("\nYouTube Metrics Relationships:")
yt_corr = corr_matrix.loc[['YoutubeViews', 'YoutubeLikes', 'YoutubeDislikes'],
                          ['YoutubeViews', 'YoutubeLikes', 'YoutubeDislikes']]
print("\nYouTube Views vs YouTube Likes:", yt_corr.loc['YoutubeViews', 'YoutubeLikes'])
print("YouTube Views vs YouTube Dislikes:", yt_corr.loc['YoutubeViews', 'YoutubeDislikes'])
print("YouTube Likes vs YouTube Dislikes:", yt_corr.loc['YoutubeLikes', 'YoutubeDislikes'])

# Budget relationships
print("\nBudget Relationships:")
for col in numeric_cols[1:]:
    corr = corr_matrix.loc['Budget', col]
    print(f"  • Budget vs {col}: {corr:.4f}")

# Final conclusion
print("\n" + "=" * 70)
print("CONCLUSION")
print("=" * 70)

# Find most correlated pair
if not max_corr.empty:
    print(f"✅ Most highly correlated: {top_pair[0]} and {top_pair[1]} (r = {max_corr.iloc[0]:.4f})")

# Check if YouTube metrics are highly correlated
if abs(corr_matrix.loc['YoutubeViews', 'YoutubeLikes']) > 0.8:
    print("✅ YouTube Views and Likes are very strongly correlated")
if abs(corr_matrix.loc['YoutubeLikes', 'YoutubeDislikes']) > 0.7:
    print("✅ YouTube Likes and Dislikes are strongly correlated")

# Check if Budget correlates with Box Office
budget_box_corr = corr_matrix.loc['Budget', 'BoxOfficeCollection']
if budget_box_corr > 0.5:
    print(f"✅ Budget moderately correlates with Box Office Collection ({budget_box_corr:.4f})")
else:
    print(f"❌ Budget has weak correlation with Box Office Collection ({budget_box_corr:.4f})")
````
